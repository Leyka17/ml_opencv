```python
img = cv.imread("woman.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python

img_res = cv.GaussianBlur(img, (21, 21), 0)
plt.imshow(img_res)
```

```python
kernel =  np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
im = cv.filter2D(img_res, -1, kernel)
fig, axs = plt.subplots(1, 2, figsize=(10, 3), constrained_layout=True)
p1 = axs[0].imshow(im)
p2 = axs[1].imshow(img_res)
plt.show()
```
