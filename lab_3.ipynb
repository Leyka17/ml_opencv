```python
import cv2 as cv
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv.imread("messi5.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
height, width = img.shape[:2]
```

```python
center = (width / 2, height / 2)
```

```python
M = cv.getRotationMatrix2D(center, 185, 1.0)
rotated = cv.warpAffine(img, M, (width, height))
plt.imshow(rotated)
```

```python
img1 = cv.imread("apple.jpg")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
h,w=img1.shape[:2]
center1=(w//2,h//2)
M = cv.getRotationMatrix2D(center1, 180, 1.0)
rotated1 = cv.warpAffine(img1, M, (w, h))
plt.imshow(rotated1)
```

```python
img2 = cv.imread("phone.png")
img = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
h,w=img2.shape[:2]
center2=(w//2,h//2)
koef_x= 0.5
koef_y = 1
M = np.float32([
    [koef_x, 0, 0],
    [0, koef_y, 0],
]
)
rotated1 = cv.warpAffine(img2, M, (int(koef_x*w), int(koef_y*h)))
plt.imshow(rotated1)
```

```python
img3 = cv.imread("skull.png")
img3 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)
plt.imshow(img3)
```

```python
rows,cols,ch = img3.shape
pts1 = np.float32([[-50,50],[200,50],[80,200]])
pts2 = np.float32([[10,100],[200,100],[200,250]])
M = cv.getAffineTransform(pts1,pts2)
dst = cv.warpAffine(img3,M,(cols,rows))[200:950,800:1370]
plt.imshow(dst)
```
